```python
import numpy as np
import pandas as pd

# Load the updated dataset
df=pd.read_csv("C:\Product_Sale_Forecasting\data\TRAIN.csv")
df.head()

```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dtypes
```

```python
df['Sales_log'] = np.log1p(df['Sales'])  # log1p = log(1 + x) to handle 0s safely
df['Order_log'] = np.log1p(df['#Order'])
df.head()
```

```python
df.info()
```

```python
# ðŸ“… 3. Date Feature Engineering
# Make sure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract useful date features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek  # Monday = 0
df['Is_Weekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >=5 else 0)
```

```python
# ðŸ†• 4. Create Additional New Features (Optional but useful)
# For example:
df['Discount_Offered'] = df['Discount'].map({'Yes': 1, 'No': 0})  # if 'Discount' is Yes/No
df['Sales_per_Order'] = df['Sales'] / (df['#Order'] + 1e-5)  # to avoid division by zero
```

```python
df.info()
```

```python
# ðŸ“Š 5. Encoding Categorical Variables
# Label Encoding for binary categorical columns
df['Discount'] = df['Discount'].map({'Yes': 1, 'No': 0})
```

```python
# One-Hot Encoding for multi-class columns
df = pd.get_dummies(df, columns=['Store_Type', 'Location_Type', 'Region_Code'], drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

Scaling Numerical Features ( Sales , Sales_log , #Order , Order_log)

```python
from sklearn.preprocessing import RobustScaler

# Select columns to scale
columns_to_scale = ['Sales', 'Sales_log', '#Order', 'Order_log']

# Initialize scaler
scaler = RobustScaler()

# Apply scaler
df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Check scaled values
df.head()
```
