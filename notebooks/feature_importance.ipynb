```python
# Step 1: Load the trained model and print features used
import pickle

# Load your trained XGBoost model
with open('C:/Product_Sale_Forecasting/models/xgboost_model.pkl', 'rb') as f:
    model = pickle.load(f)

# Get and print feature names the model was trained on
features_used = model.feature_names_in_
print("Features used by the model:")
print(features_used)

# Step 2: Plot Feature Importances
import matplotlib.pyplot as plt
import pandas as pd

# Get feature importances from the model
importances = model.feature_importances_

# Combine feature names and their importance into a DataFrame
importance_df = pd.DataFrame({
    'Feature': features_used,
    'Importance': importances
})

# Sort by importance
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Most important at top
plt.title("XGBoost Feature Importances")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```
